**Conclusões**

Por fim, conseguimos chegar a algumas conclusões quanto a cada categoria:

Idade: se relaciona com a renda à medida que aumenta, até as pessoas começarem a se aposentar.

Classe de trabalho: pessoas que trabalham pro governo federal e empreendedores vinculados à empresas tem uma renda maior.

Educação: quanto maior os níveis de educação maior a renda, apesar da maioria dos trabalhadores terem apenas o ensino médio completo.

Número de anos dedicados à educação: quanto mais anos dedicados, maior a renda, porém ensinos incompletos não são tão relevantes.

Estado civil: pessoas casadas com cônjuge presente tem uma renda maior, devido à necessidade de sustentar outras pessoas além de si próprias, incluindo filhos.

Ocupação: especialidades profissionais e cargos de gestão executivas tem uma renda maior, porém a distruição de bem diversas quanto às diferentes ocupações.

Relacionamento: maridos e esposas tendem a ter uma renda maior, pelo mesmo motivo do estado civil, ou seja, a necessidade de ter uma renda maior para sustentar a família.

Etnia: pessoas de origem branca e asiática tendem a ter uma renda maior, e pessoas de origem negra tendem a ter uma renda menor.

Gênero: homens tendem a ganhar mais que mulheres, apesar de mulheres trabalharem menos horas por semana, devido à maternidade e muitas vezes uma responsabilidade doméstica maior. Apesar disso, não deixa de ser problemática a diferença de renda para mesmas posições.

Ganho de investimentos externos: pessoas com renda maior tendem a ganhar mais por investimentos externos, visto que tem mais dinheiro para investir, além de pessoas com maior idade terem um ganho maior pois dependem muitas vezes dessa aposentadoria.

Horas por semana: pessoas que trabalham mais tendem a ter uma renda maior, porém as horas por semana seguem um padrão de 40 na maioria dos casos.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

%matplotlib inline
```

```python
lr = LogisticRegression(solver='liblinear')
dados_dummies = pd.concat([dados['age'], pd.get_dummies(dados['workclass'])], axis=1)
dados_dummies = pd.concat([dados_dummies, dados['educational-num']], axis=1)
dados_dummies = pd.concat([dados_dummies, pd.get_dummies(dados['marital-status'])], axis=1)
dados_dummies = pd.concat([dados_dummies, pd.get_dummies(dados['occupation'])], axis=1)
dados_dummies = pd.concat([dados_dummies, pd.get_dummies(dados['relationship'])], axis=1)
dados_dummies = pd.concat([dados_dummies, pd.get_dummies(dados['race'])], axis=1)
dados_dummies = pd.concat([dados_dummies, pd.get_dummies(dados['gender'])], axis=1)
dados_dummies = pd.concat([dados_dummies, dados['capital-gain']], axis=1)
dados_dummies = pd.concat([dados_dummies, dados['capital-loss']], axis=1)
dados_dummies = pd.concat([dados_dummies, dados['hours-per-week']], axis=1)
x = dados_dummies
y = pd.get_dummies(dados['income'])['>50K']
scores = cross_val_score(lr, x, y, cv=10, scoring='accuracy')
print('Acurácia média:', scores.mean())
```

```python
age = input('Idade:')
workclass = input('Classe de trabalho:')
educational_num = input('Número de anos dedicados à educação:')
marital_status = input('Estado-civil:')
occupation = input('Ocupação:')
relationship = input('Relacionamento familiar:')
race = input('Etnia:')
gender = input('Gênero:')
capital_gain = input('Ganho em investimentos externos:')
capital_loss = input('Perda em investimentos externos:')
hours_per_week = input('Horas trabalhadas por semana:')
pessoa = {'age': age, 'workclass': workclass, 'fnlwgt': 0, 'education': 0, 'educational-num': educational_num, 'marital-status': marital_status, 'occupation': occupation, 'relationship': relationship, 'race': race, 'gender': gender, 'capital-gain': 0, 'capital-loss' : 0, 'hours-per-week': hours_per_week, 'native-country': 0, 'income': 0}
dados.append(pessoa, ignore_index = True)
dados_sdummies = pd.concat([dados['age'], pd.get_dummies(dados['workclass'])], axis=1)
dados_sdummies = pd.concat([dados_sdummies, dados['educational-num']], axis=1)
dados_sdummies = pd.concat([dados_sdummies, pd.get_dummies(dados['marital-status'])], axis=1)
dados_sdummies = pd.concat([dados_sdummies, pd.get_dummies(dados['occupation'])], axis=1)
dados_sdummies = pd.concat([dados_sdummies, pd.get_dummies(dados['relationship'])], axis=1)
dados_sdummies = pd.concat([dados_sdummies, pd.get_dummies(dados['race'])], axis=1)
dados_sdummies = pd.concat([dados_sdummies, pd.get_dummies(dados['gender'])], axis=1)
dados_sdummies = pd.concat([dados_sdummies, dados['capital-gain']], axis=1)
dados_sdummies = pd.concat([dados_sdummies, dados['capital-loss']], axis=1)
dados_sdummies = pd.concat([dados_sdummies, dados['hours-per-week']], axis=1)
lr.fit(x, y)
print('Previsão: (1 para salário maior que 50K, 0 para salário menor que 50K)', lr.predict(dados_sdummies.iloc[-1:]))
```
